```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import math
from scipy import signal
from scipy.fftpack import fft
from scipy import ndimage
from scipy import special
%matplotlib inline
```

```python
#Gain control function
def gain_func(v):
    if v <0:
        print('廢物')
        return 1
    else:
        return 1/(1+v**4)
```

```python
def K(x,s):
    return(np.exp(-(x/s)**2))/s/np.sqrt(np.pi)
```

# delta function

```python
nstep = 3000
frequency = 8#Hz
dt = 0.001
nperiod = int((1/frequency)/dt)

# generation of stimulations
s = np.zeros(nstep)
for i in range(nstep):
    if i % nperiod ==3:
        s[i] = 1
#s = np.convolve(tp,s,'same')*dt
plt.plot(s)
```

```python
x = np.zeros_like(s)#y output
y = np.zeros_like(s)#y output
B = 35
C = 1
beta = 17
for i in range(len(s)):
    if s[i] == 1:
        y[i] = np.exp(-beta*nperiod*dt)*(B*x[i-nperiod] + y[i-nperiod])
        x[i] = C/(1+y[i]**4)
```

```python
plt.subplot(311)
plt.plot(s)
plt.title('Stimulus')
plt.subplot(312)
plt.plot(x)
plt.title('Output')
plt.subplot(313)
plt.plot(y)
plt.title('feedback variable')
plt.gcf().set_size_inches(12,6)
```

```python
x = np.zeros_like(s)#y output
y = np.zeros_like(s)#y output
gains = np.zeros_like(s)#y output

B = 35
tau = 1/17

v = 0.0
for i in range(len(s)):
    feedback_filter = np.exp(-np.arange(0,i)*dt/tau)
    y[i]= B*np.dot(x[0:i],feedback_filter[::-1])
    gains[i] = gain_func(y[i]).copy()
    x[i] = gains[i]*s[i]
```

```python
plt.subplot(311)
plt.plot(s)
plt.title('Stimulus')
plt.subplot(312)
plt.plot(x)
plt.title('Output')
plt.subplot(313)
plt.plot(gains)
plt.title(' feedback variable')
plt.gcf().set_size_inches(12,6)
```

# square pulse

```python
nstep = 3000
frequency = 8#Hz
nperiod = int((1/frequency)/dt)
# generation of stimulations
s = np.zeros(nstep)
for i in range(nstep):
    if i % nperiod ==0:
        s[i:i+62] = 1
#s = np.convolve(tp,s,'same')*dt
plt.plot(s)
```

```python
x = np.zeros_like(s)#y output
y = np.zeros_like(s)#y output
gains = np.zeros_like(s)#y output

B = 35
tau = 0.04

v = 0.0
for i in range(len(s)):
    feedback_filter = np.exp(-np.arange(0,i)*dt/tau)
    y[i]= B*np.dot(x[0:i],feedback_filter[::-1])
    gains[i] = gain_func(y[i]).copy()
    x[i] = gains[i]*s[i]
```

```python
plt.subplot(411)
plt.plot(s)
plt.title('Stimulus')
plt.subplot(412)
plt.plot(x*s)
plt.title('Output')
plt.subplot(413)
plt.plot(gains)
plt.title(' gain')
plt.subplot(414)
plt.plot(y)
plt.title('feedback varible')
plt.gcf().set_size_inches(12,10)
```

# sin wave

```python
plt.subplot(311)
plt.plot(s)
plt.title('Stimulus')
plt.subplot(312)
plt.plot(x*s)
plt.title('Output')
plt.subplot(313)
plt.plot(gains)
plt.title(' gain')
plt.gcf().set_size_inches(12,6)
```
