# Functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.optimize import curve_fit
import os
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#This function deals with differentiation of angle within [-pi,pi).
#It will fail if angle array x change too fast.
def angular_gradient(x):
    diffx= np.gradient(x)
    diffx[np.abs(diffx)>0.6*np.pi] -= np.pi*diffx[np.abs(diffx)>0.6*np.pi]/abs(diffx[np.abs(diffx)>0.6*np.pi])
    return diffx
```

# main

```python
fig, ax = plt.subplots()
exp_folder = 'D:\\GoogleDrive\\retina\\Chou\'s data\\20210504'
annots = loadmat(exp_folder+'\\merge\\merge_0224_cSTA_wf_3min_Q100_re', squeeze_me = True)
x = annots['bin_pos']
x = x-np.mean(x)
dt = 1/60
cn = 38-1

spike = annots['reconstruct_spikes'][cn]
rstate, _ = np.histogram(spike, np.arange(len(x)+1)*dt)
cSTA = np.correlate(x , rstate, 'same')/ np.correlate(np.ones_like(x), rstate, 'same')
cSTA = cSTA[int(len(cSTA)/2):int(len(cSTA)/2-1/dt)-1:-1]
taxis = -np.arange(len(cSTA))*dt
ax.plot(taxis, cSTA, 'b+:')
```

# NGD1L 

```python
def NGD_TK(T, alpha, beta, g, tau_y, Amp):
    dt = np.mean(np.diff(T))
    dt = 0.1*dt
    T = np.arange(0, T[-1]+dt, dt)
    x = T/tau_y /tau_y *np.exp(-T/tau_y)*dt
    
    y=np.zeros_like(x)
    z=np.zeros_like(x)
    for j in range(len(x)-1):
        dy=dt*(-alpha*y[j]+(x[j]-z[j]))
        dz=dt*(-beta*z[j]+g*y[j])
        y[j+1]=y[j]+dy
        z[j+1]=z[j]+dz

    nTK = y[::10]/(dt) #normalized by integration
    return -Amp*nTK
```

```python
beta = 1.2
alpha = 25.
g = 600.
tau_y=0.03
popt,pcov = curve_fit(NGD_TK, np.abs(taxis), cSTA, p0=[alpha, beta, g, tau_y, 5], bounds = (0, np.inf))

fit_cSTA = NGD_TK(np.abs(taxis), *popt).copy()
error = sum((fit_cSTA-cSTA)**2)

[alpha, beta, g, tau_y, Amp] = popt
ax.plot(taxis, fit_cSTA, 'r-')
fig
```

```python
popt
```

```python
domega = 0.01
omega = np.arange(-domega,10,domega) #limited by [pi/dt,pi/dt)
TKdw = np.sum(fit_cSTA[np.newaxis,:]*np.exp(-1j*np.abs(taxis[np.newaxis,:])*omega[:,np.newaxis]), axis = 1)*dt

phi = np.angle(TKdw)
delta = -angular_gradient(phi)/domega
plt.plot(omega, delta)
plt.axhline(0,linewidth=0.5, linestyle='--')
plt.ylabel(r'$\delta$')
plt.xlabel(r'Angular Frequency ($s^{-1}$)')
```

# NGD3L

```python
def NGD3L_TK(T, tauC, tauH, tauB, tauA, tauG,
             a, b, c, d, e, f, h, i, tau_y, Amp):
    dt = np.mean(np.diff(T))
    dt = 0.1*dt
    T = np.arange(0, T[-1]+dt, dt)
    x = T/tau_y /tau_y *np.exp(-T/tau_y)*dt
    
    C=np.zeros_like(x)
    H=np.zeros_like(x)
    B=np.zeros_like(x)
    A=np.zeros_like(x)
    G=np.zeros_like(x)
    for j in range(len(x)-1):
        dC=dt*(-C[j]/tauC + (x[j]-i*H[j]))
        dH=dt*(-H[j]/tauH + a*C[j])
        dB=dt*(-B[j]/tauB + b*C[j]-c*H[j]-d*A[j] )
        dA=dt*(-A[j]/tauA + e*B[j])
        dG=dt*(-G[j]/tauG + f*B[j]-h*A[j])
        
        C[j+1]=C[j]+dC
        B[j+1]=B[j]+dB
        H[j+1]=H[j]+dH
        A[j+1]=A[j]+dA
        G[j+1]=G[j]+dG
    nTK = G[::10]/(dt)/100 #normalized by integration
    return -Amp*nTK

def plotNGD3L_TK(T, tauC, tauH, tauB, tauA, tauG,
                 a, b, c, d, e, f, h, i, tau_y, Amp):
    dt = np.mean(np.diff(T))
    dt = 0.1*dt
    T = np.arange(0, T[-1]+dt, dt)
    x = T/tau_y /tau_y *np.exp(-T/tau_y)*dt
    
    C=np.zeros_like(x)
    H=np.zeros_like(x)
    B=np.zeros_like(x)
    A=np.zeros_like(x)
    G=np.zeros_like(x)
    for j in range(len(x)-1):
        dC=dt*(-C[j]/tauC + (x[j]-i*H[j]))
        dH=dt*(-H[j]/tauH + a*C[j])
        dB=dt*(-B[j]/tauB + b*C[j]-c*H[j]-d*A[j] )
        dA=dt*(-A[j]/tauA + e*B[j])
        dG=dt*(-G[j]/tauG + f*B[j]-h*A[j])
        
        C[j+1]=C[j]+dC
        B[j+1]=B[j]+dB
        H[j+1]=H[j]+dH
        A[j+1]=A[j]+dA
        G[j+1]=G[j]+dG

    plt.plot(T,C/max(C), label='C')
    plt.plot(T,B/max(B), label='B')
    plt.plot(T,G/max(G), label='G')
    plt.legend()
    
# plotNGD3L_TK(np.abs(taxis), 1/alpha, 1/beta, 1/gamma, 1/beta, 1/beta, a, b, c, 0, 10, 10, 0,tau_y, Amp)
```
